```python
gudang = []

def muat_data():
    try:
        with open("data_gudang.txt", "r") as file:
            baris_baris = file.readlines()
            for baris in baris_baris:
                # .strip() membuang enter (\n) agar teks bersih
                gudang.append(baris.strip())
    except FileNotFoundError:
        # Jika file belum ada (pertama kali dijalankan), abaikan saja
        pass

# --- FUNGSI 2: TAMPILAN MENU ---
def tampilkan_menu():
    print("\n=== Menu Utama")
    print("1. Masuk Barang")
    print("2. Cek Stok Barang")
    print("3. Hapus barang")
    print("4. Tutup Program / Keluar")

# --- FUNGSI 3: TAMBAH BARANG (KE RAM & HARDDISK) ---
def tambah_barang():
    barang_baru = input("Masukan barang : ")
    
    # Simpan ke List (RAM)
    gudang.append(barang_baru)
    
    # Simpan ke File (Harddisk) - Mode "a" (Append)
    with open("data_gudang.txt", "a") as file:
        file.write(barang_baru + "\n")
        
    print(barang_baru, "sudah ditambahkan permanen")

# --- FUNGSI 4: LIHAT BARANG ---
def lihat_barang():
    print("Stok gudang : ", gudang)

# --- FUNGSI 5: HAPUS BARANG (DARI RAM) ---
def hapus_barang():
    barang_keluar = input("Masukan barang yang akan dihapus: ")
    if barang_keluar in gudang:
        gudang.remove(barang_keluar)
        print(barang_keluar, " berhasil dihapus")
    else:
        print("Barang tidak ditemukan, silahkan cek stok dulu.")

# ==========================================
#PROGRAM UTAMA (MAIN LOOP)
# ==========================================

# 1. Panggil ingatan masa lalu dulu
muat_data()

print("Sistem Gudang Jarvis Aktif...")

while True :
    tampilkan_menu()
    
    pilihan = input("Pilih menu 1-4 : ")

    if pilihan == "1":
        tambah_barang()
    
    elif pilihan == "2":
        lihat_barang()
    
    elif pilihan == "3":
        hapus_barang()

    elif pilihan == "4":
        print("Sampai jumpa! Data Anda aman.")
        break

    else : 
        print("Pilihan salah, masukkan angka 1, 2, 3, atau 4")
```

```python
daftar_hadir =[]

def baca_data_absensi():
    try:
        with open ("absensi.txt", "r") as file:
            semua_baris = file.readlines()
            for baris in semua_baris :
                daftar_hadir.append(baris.strip())
    except FileNotFoundError:
        pass
def tampilan_menu():
    print("1. Catat Kehadiran")
    print("2. Lihat Siapa Yang Hadir")
    print("3. Hapus Kehadiran") 
    print("4. Keluar/Selesai Absensi")



def catat_kehadiran():
    nama_siswa = input ("Nama Siswa")
    daftar_hadir.append(nama_siswa)
        
    with open("absensi.txt", "a") as file:
        file.write(nama_siswa +"\n")
        print(nama_siswa, "hadir")


def siswa_hadir():
    print("siswa yang hadir ", daftar_hadir)

def hapus_kehadiran():
    tidak_hadir = input("Nama Siswa")
    if tidak_hadir in daftar_hadir :
        daftar_hadir.remove(tidak_hadir)
        print(tidak_hadir, "tidak hadir")
    else : 
        print("siswa tidak dalam daftar hadir")


baca_data_absensi()
while True:
    tampilan_menu()

    pilihan = input("pilih 1-4")

    if pilihan == "1":
        catat_kehadiran()

    elif pilihan == "2":
        siswa_hadir()

    elif pilihan == "3":
        hapus_kehadiran()

    elif pilihan == "4":
        print("Anda Keluar")
        break

    else :
        print("masukan 1-4")
    
            
        
```
