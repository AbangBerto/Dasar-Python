```python
gudang = []

def tampilkan_menu():
    print("\n=== Menu Utama")
    print("1. Masuk Barang")
    print("2. Cek Stok Barang")
    print("3. Hapus barang")
    print("4. Tutup Program / Keluar")

# --- BAGIAN LOOPING ---
while True:
    # --- PANGGIL FUNGSI ---
    # Cukup 1 baris ini menggantikan 4 baris print
    tampilkan_menu() 

    # --- BAGIAN BAWAH INI TIDAK ADA YANG BERUBAH ---
    # (Sama persis seperti kode Anda sebelumnya)
    pilihan = input("pilih 1/4 : ")

    if pilihan == "1":
        gudang_baru = input("Masukan Barang: ")
        gudang.append(gudang_baru)
        print("Barang berhasil ditambahkan")

    elif pilihan == "2":
        print("Stok Gudang: ", gudang)
    
    elif pilihan == "3":
        barang_keluar = input("Masukan barang yang ingin dihapus: ")
        if barang_keluar in gudang:
            gudang.remove(barang_keluar)
            print("Barang berhasil dihapus")
        else:
            print("Barang tidak ada di gudang")
    
    elif pilihan == "4":
        print("Anda berhasil keluar")
        break 
    
    else:
        print("Pilihan tidak ada")
```

```python
gudang = []
def tampilkan_menu():
    print("\n=== Menu Utama")
    print("1. Masuk Barang")
    print("2. Cek Stok Barang")
    print("3. Hapus barang")
    print("4. Tutup Program / Keluar")

def tambah_barang():
    barang_baru = input("Masukan barang : ")
    gudang.append(barang_baru)
    print(barang_baru, "sudah ditambahkan")
    
def lihat_barang():
    print("Stok gudang : ", gudang)

def hapus_barang():
    barang_keluar = input("Masukan barang yang akan dihapus")
    if barang_keluar in gudang:
        gudang.remove(barang_keluar)
        print(barang_keluar, " berhasil dihapus")

    else:
        print("Barang tidak ditemukan silahkan cek stok untuk melihat barang anda")

while True :
    tampilkan_menu()
    
    pilihan = input("Pilih antara 1-4 : ")
    if pilihan == "1":
        tambah_barang()
    
    elif pilihan == "2":
        lihat_barang()
    
    elif pilihan == "3":
        hapus_barang()

    elif pilihan == "4":
        break

    else : 
        print("Masukan 1-4")
```
