```python
gudang = []

while True:
    print("\n=== Menu Utama")
    print("1. Masuk Barang")
    print("2. Cek Stok Barang")
    print("3. Hapus barang")
    print("4. Tutup Program / Keluar")

    pilihan = input("pilih 1/4 : ")

    if pilihan == "1":
        gudang_baru = input("Masukan Barang: ")
        gudang.append(gudang_baru)
        print("Barang berhasil ditambahkan ke Gudang")

    elif pilihan == "2":
        print("Stok Gudang: ", gudang)
    
    elif pilihan == "3":
        barang_keluar = input("Masukan barang yang ingin dihapus: ")
        if barang_keluar in gudang:
            gudang.remove(barang_keluar)
            print("barang berhasil dihapus")
        else:
            print("barang tidak ada di gudang")
    
    elif pilihan == "4":
        print("Anda berhasil keluar")
        break 
    
    else:
        print("Pilihan tidak ada masukan 1, 2, 3, atau 4")
```
