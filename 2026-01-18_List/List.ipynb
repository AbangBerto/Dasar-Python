```python
# Membuat List
database_ai = ["Jarvis", 1.0, "Jawa Tengah", "Online"]

# Menampilkan seluruh isi lemari
print(database_ai)
```

```python
# di sini kita bisa mengambil data dari list di atas misal kita ingin mgambil lokasi 
# ingat di program index mulai dari 0 
# lokasi berarti di index 2
```

```python
# kita juga bisa menambahkan isi dari list di atas dengan perentah .append
database_ai.append("Baterai 100%")
                   
```

```python
# sekarang kita cek isi listnya
print(database_ai)
```

```python
# kita sudah belajar menambahkan list, lalu bagaimana jika ingin menghapusnya?
database_ai.remove("Jawa Tengah")
print(database_ai)
```

```python
#kita juga bisa menghapus melalui indexnya dengan printah .pop()
database_ai.pop(0)
print(database_ai)
```
