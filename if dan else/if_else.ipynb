```python
# dari output program di atas dia tidak memenuhi kondisi maka dia akan menjalankan perintah else: 

# jadi Saat program berjalan dan sampai di baris if, dia berhenti sejenak untuk menimbang angka kecepatan yang Anda masukkan. 
# Jika angkanya memang di bawah 80 (Benar), komputer akan masuk ke dalam blok if untuk menampilkan peringatan, lalu secara otomatis 
# membuang/mengabaikan seluruh bagian else. Sebaliknya, jika angkanya ternyata 80 atau lebih (Salah),
# komputer akan langsung melompati blok if tersebut dan hanya menjalankan perintah yang ada di dalam else. Intinya, 
# komputer hanya akan memilih satu jalan saja, tidak pernah keduanya sekaligus.
```

```python
#lalu bagaimana jika ada 3 condisi? 
# apakah pakai if, if, else? 
# nah kita coba saja

# dari ouput di atas kita tahu bahwa setelah melaui pemeriksaan dia di bawah 80 akan blok if karena kecepatannya 10. nah sekarang bagaimana jika kita
# memasukan kecepatannya >80? 
# oke perthatikan di output di bawah ini 
 
nama_ai = input ("Nama AI : ")
versi = float (input ("versi : "))
kecepatan = float (input("kecepatan : "))

print("Versi : " , versi)
print("Kecepatan : " , kecepatan)

versi_baru = versi + 1.0 
print("sedang menginstall pembaruan.....")
print("Versi AI anda sekarang menjadi : ", versi_baru)

print("-" * 20) #jangan bingung ini apa ini hanya spasi biar rapi 

if kecepatan > 100 :
    print("Bahaya! terlalu cepat hati-hati bisa overhead")

if kecepatan < 80 : 
    print("Peringatan : Sistem berjalan lambat")
   
else  : 
    print("Sistem berjalan Optimal")
```

```python
# Pehatikan muncul masalah munculnya dua output sekaligus padahal kita ingin memunculkan  yang > 100 saja, ini  terjadi karena 
# penggunaan dua if secara terpisah membuat program menganggap ada dua pemeriksaan yang berdiri sendiri dan tidak saling berhubungan
# ("tidak sepaket"). Saat Anda memasukkan angka 150, pemeriksaan pertama (apakah kecepatan > 100) berhasil, sehingga program mencetak pesan "Bahaya". 
# Namun, program tidak berhenti di situ;ia tetap lanjut melakukan pemeriksaan kedua (apakah kecepatan < 80). Karena angka 150 jelas tidak kurang dari 80, 
# maka pemeriksaan kedua ini dianggap gagal (False). Akibatnya, blok else—yang tugasnya memang aktif saat if di atasnya
# gagal—langsung mengambil alih dan mencetak "Sistem Optimal". Singkatnya, else tersebut "buta" dan tidak peduli bahwa angka
# itu sebenarnya sudah terdeteksi bahaya di pemeriksaan pertama, karena ia hanya setia pada if yang tepat berada di atas kepalanya saja. 

    
# Bagaimana jika kita hanya ingin mengeluarkan salah satu saja? dalam contoh yang > 100 saja.
# caranya kita buat sepaket dengan menggunakan (elif) dia akan memperdulikan pemeriksaan sebelumnya
```

```python
nama_ai = input ("Nama AI : ")
versi = float (input ("versi : "))
kecepatan = float (input("kecepatan : "))

print("Versi : " , versi)
print("Kecepatan : " , kecepatan)

versi_baru = versi + 1.0 
print("sedang menginstall pembaruan.....")
print("Versi AI anda sekarang menjadi : ", versi_baru)

print("-" * 20) #jangan bingung ini apa ini hanya spasi biar rapi 

if kecepatan > 100 :
    print("Bahaya! terlalu cepat hati-hati bisa overhead")

elif kecepatan < 80 : 
    print("Peringatan : Sistem berjalan lambat")
   
else  : 
    print("Sistem berjalan Optimal")
```
